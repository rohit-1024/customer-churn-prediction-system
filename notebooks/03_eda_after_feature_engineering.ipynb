## 📊 EDA Summary (After Feature Engineering)

### Dataset Overview
- After performing Feature Engineering, the dataset has been transformed into a **fully machine learning–ready format**.
- All categorical variables have been converted into **numeric representations**, and additional derived features have been created to improve the predictive capability of the model.
- The dataset is now structured so that **machine learning algorithms can directly process all features**.

### Feature Engineering Results
- The **target variable (`churn`)** was converted from categorical values to **binary numeric format**.

        Yes → 1  
        No  → 0

- Binary categorical features such as `partner`, `dependents`, `phone_service`, and service-related columns were converted into **0/1 numeric values**.
- Multi-category columns such as `internet_service`, `contract`, and `payment_method` were encoded using **One-Hot Encoding**.
- One-Hot Encoding created separate binary columns for each category, allowing machine learning models to interpret them correctly.

### Engineered Features Created
To capture deeper behavioral patterns, additional derived features were created:

1. **Tenure Group Feature (`tenure_group`)**
   - Customers were grouped into lifecycle stages based on the number of months they stayed with the company.
   - Example groups include:
   
        0–12 months   → new customers  
        12–24 months  → early stage customers  
        24–48 months  → mid-term customers  
        48+ months    → loyal customers  

   - This feature helps identify churn patterns across different customer lifecycle stages.

2. **Service Count Feature (`services_count`)**
   - Several columns represent optional services subscribed by customers.
   - A new feature was created by counting the number of services used by each customer.
   - Customers using **more services generally show stronger engagement and lower churn probability**.

3. **Average Monthly Spend (`avg_monthly_spend`)**
   - A derived feature was created using:

        avg_monthly_spend = total_charges / tenure

   - This captures the **spending behavior of customers** and helps identify customers who may churn after short high-spending periods.
