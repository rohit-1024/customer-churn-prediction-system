# 🔍 Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing and understanding the dataset before building machine learning models.  
It helps identify data quality issues, understand feature distributions, and discover patterns that may influence the prediction of **customer churn**.

The following steps are performed during the EDA phase:

---

## 📂 1. Load Dataset
- Import the dataset into the environment.
- Verify that the data loads correctly.

---

## 👀 2. Inspect Initial Records
- View the **first few rows** using `head()`.
- View the **last few rows** using `tail()`.
- Helps understand the basic structure of the dataset.

---

## 📏 3. Dataset Shape
- Determine the **number of rows and columns** in the dataset.
- Helps understand dataset size.

---

## 🏷️ 4. Column Names
- List all column names in the dataset.
- Helps identify available features.

---

## 🧩 5. Data Types
- Check data types of each column.
- Identify:
  - Numerical features
  - Categorical features
  - Date/time features

---

## ℹ️ 6. Dataset Information
- Display dataset metadata using `info()`.
- Includes:
  - Column types
  - Non-null values
  - Memory usage

---

## 📊 7. Summary Statistics
- Generate descriptive statistics using `describe()`.
- Understand:
  - Mean
  - Median
  - Standard deviation
  - Min / Max values
  - Quartiles

---

## ❓ 8. Missing Values
- Identify missing values in each column.
- Determine whether data cleaning or imputation is required.

---

## 🔁 9. Duplicate Records
- Check if duplicate rows exist in the dataset.

---

## ⚠️ 10. Invalid Values
- Detect impossible or unrealistic values such as:
  - Negative values
  - Zero values where not allowed
  - Out-of-range values

---

## 🔢 11. Unique Values in Categorical Features
- Identify unique categories in categorical columns.
- Helps understand possible feature values.

---

## 📉 12. Category Distributions
- Analyze the frequency distribution of categorical variables.
- Helps detect class imbalance.

---

## 📈 13. Numerical Feature Distributions
- Analyze distribution of numeric variables using:
  - Histograms
  - Box plots
  - Density plots

---

## 📅 14. Date Range Analysis (if applicable)
- Check minimum and maximum values of date columns.
- Understand time coverage of the dataset.

---

## 💾 15. Memory Usage
- Evaluate memory consumption of the dataset.

---

## 🎯 Purpose of EDA in This Project

The goal of EDA in this **Telecom Customer Churn Prediction System** is to:

- Understand customer behavior patterns
- Detect data quality issues
- Identify important features affecting churn

## Key Data Issues & Cleaning Rules :

During Exploratory Data Analysis (EDA), several data quality issues and preprocessing requirements were identified. These must be addressed before building machine learning models.

---

### 1. Identifier Column

The column **`customerID`** uniquely identifies each customer.

- It does **not provide predictive value** for churn prediction.
- Identifiers should **not be used as ML features**.

**Cleaning Rule**

Remove the identifier column during preprocessing.

```python
df.drop("customerID", axis=1, inplace=True)
```

---

### 2. Incorrect Data Type

The column **`TotalCharges`** is stored as **object (string)** instead of a numeric type.

This occurs because some rows contain **blank values**, preventing automatic numeric conversion.

**Cleaning Rule**

Convert `TotalCharges` to a numeric data type.

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

---

### 3. Missing Values

After converting `TotalCharges` to numeric, blank values become **NaN (missing values)**.

These missing values must be handled before model training.

**Cleaning Rule**

Remove rows containing missing values.

```python
df = df.dropna()
```

---

### 4. Inconsistent Feature Representation

Most categorical features use **"Yes / No"** values, but the column **`SeniorCitizen`** uses **0 / 1**.

This creates inconsistency in categorical representation.

**Cleaning Rule**

Convert `SeniorCitizen` to a categorical format (**"Yes / No"**).

---

### 5. Column Naming Consistency

Some column names do not follow consistent naming conventions.

Examples:

- `tenure` → `Tenure`
- `gender` → `Gender`

Consistent naming improves dataset readability and maintainability.

**Cleaning Rule**

Standardize column names.

---

### 6. Class Imbalance in Target Variable

The target column **`Churn`** is **imbalanced**.

Approximate distribution:

- **~73% → No Churn**
- **~27% → Churn**

Imbalanced datasets require careful model evaluation.

**Handling Strategy**

Use evaluation metrics beyond accuracy, such as:

- Precision
- Recall
- F1 Score
- ROC-AUC

---

### Summary of Data Cleaning Steps

Based on EDA findings, the following preprocessing steps will be applied:

1. Remove identifier column (`customerID`)
2. Convert `TotalCharges` to numeric
3. Convert `SeniorCitizen` to categorical (Yes / No)
3. Handle missing values created during conversion
4. Ensure consistent feature representation
5. Standardize column naming conventions
6. Address class imbalance during model evaluation

These steps prepare the dataset for **feature engineering and machine learning model training**.
