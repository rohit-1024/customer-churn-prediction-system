```python

# CELL 19 — Group columns by their types (Numerical / Categorical)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

print("\nNumerical Columns:", numeric_cols[:10], "\n\nTotal Numeric Columns:", len(numeric_cols))
print("\nCategorical Columns:", categorical_cols[:10], "\n\nTotal Categorical Columns:", len(categorical_cols),"\n")
```

```python

# CELL 20 — Numerical feature's distribution plots

for feature in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python

# CELL 21 — Unique values (categories / classes) in categorical columns

for col in categorical_cols:
    print(f"\n --> Unique values in column '{col}' : {df[col].unique()}")
print()
```

```python

# CELL 22 — Category / class distribution in each categorical column

print("\n----------------------------------------------\n")
for col in categorical_cols:
    print(df[col].value_counts())
    print("\n----------------------------------------------\n")
print()
```

```python

# CELL 23 — Show all the categorical feature columns

print(f"\nCategorical Columns : \n\n{categorical_cols}\n")
```

```python

# CELL 24 — Plots for each -> 'Categorical Feature' vs 'Churn'

# NOTE: Except 'customerID' and 'TotalCharges' columns as they are not actually categorical features

print("\n-------------------------------------------------------------------------------------\n")
for col in categorical_cols:    
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue="churn", data=df)
    
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.show()
    print("\n-------------------------------------------------------------------------------------\n")
```

```python

# CELL 25 — Draw / plot the Correlation Matrix

print("\n-------------------------------------------------------------------------------------\n")
corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix") 
plt.show()
print("\n-------------------------------------------------------------------------------------\n")
```

## 📊 EDA Summary (After Data Cleaning)

### Dataset Overview
- The dataset contains telecom customer information used to analyze **customer churn behavior**.
- After data preprocessing, the dataset is **clean, consistent, and ready for the next stages of the ML pipeline**.

### Data Cleaning Results
- The **identifier column (`customer_id`)** was removed since it does not contribute to churn prediction.
- The **`total_charges` column** was converted from *object* to **numeric (float)** data type.
- Rows containing **missing values in `total_charges`** were removed.
- Duplicate records were checked and removed to maintain **data integrity**.
- Column names were standardized to **snake_case** for consistency across the project.

### Key Observations from Clean Data
1. The dataset contains **~7010 telecom customers** with multiple service-related features.
2. The **target variable (`churn`) is imbalanced**, with fewer churned customers compared to non-churned customers.
3. Customers with **month-to-month contracts** show higher churn rates than those with longer-term contracts.
4. Customers with **shorter tenure** tend to churn more frequently.
5. Higher **monthly_charges** are associated with an increased probability of churn.

### Dataset Readiness
After cleaning and verification:
- All columns have **correct data types**.
- **Missing values and duplicates** have been handled.
- Column naming is **standardized and consistent**.

The dataset is now **clean and ready for the next stage of the ML pipeline: Feature Engineering**, where categorical encoding, feature transformation, and preparation for model training will be performed.
